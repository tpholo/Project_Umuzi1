```python
import pandas as pd 
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
file = pd.read_excel('PCspecs.xlsx')
file  
```

```python
def priceToInt(string):
    return float(string.replace(',', ''))
```

```python
#Creating new column New_Price and change the values 
#list comprehension
file['New_Price'] = [priceToInt(row) for row in file['Price (ZAR)']]
file['New_Price'] 
```

```python
#Using the rename method to change the colums names
file = file.rename(columns = {'Screen size (")' : 'Screen_Size','Hard drive':'Hard_drive' , 'Optical drive' : 'Optical_drive' , 'Battery cells': 'Battery_cells'})
file
```

```python
#Creating a new column and New-Hard_Drive with numbers only from coluumn Hard_drive
file['New_Hard_drive'] = file['Hard_drive'].str.extract( '(\d+)',expand=True)
file['New_Hard_drive']
```

```python
#In column Hew_Hard_drive we change the values that are less than 10 by multiplying them with 1000
#to get the value of TB - GB 
file['New_Hard_drive'] = [int(x) * 1000 if (int(x) < 10) else x for x in file['New_Hard_drive'] ]
file
```

```python
#Now we change the values of RAM column to only int by extracting the strings 
file['New_Ram'] = file['RAM'].str.extract( '(\d)',expand=True)
file['New_Ram']
```

```python
file['New_Ram'] = file['New_Ram'].replace(np.nan , 8, regex=True)
file['New_Ram']
```
